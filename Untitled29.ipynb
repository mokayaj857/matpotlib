```python
import matplotlib.pyplot as plt
names= (1,2,3,4,5,7)
year =(2001,2003,2009,2004,2007,2010)
plt.scatter(names, year, linewidth=2, marker='o' ,linestyle='dashed',label='my_journey')
plt.xlabel('names')
plt.ylabel('year')
plt.title('my_findings')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define 'views' as a list of data points
views = [100, 200, 400, 599,544,345,345]
time =range(1,8)

# Plot the views
plt.plot(views,time,label='youtube views',color='indigo', marker='o',linestyle='dashdot',linewidth=1)
plt.xlabel('time')
plt.ylabel('Views')

plt.legend()
plt.show()
```
